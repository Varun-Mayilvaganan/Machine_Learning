```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Label encoding of breast cancer datasets
cancer_data = pd.read_csv('/content/breast-cancer.csv')
```

```python
cancer_data.head()
```

```python
cancer_data.shape
```

```python
# finding the count of catrgories
cancer_data['diagnosis'].value_counts()
```

```python
# load the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataframe
cancer_data['outcome'] = labels
cancer_data
```

```python
# 0 represents benign
# 1 represents Malignant
cancer_data['outcome'].value_counts()
```

```python
iris_data = pd.read_csv('/content/iris.data.csv')
iris_data.head()
```

```python
iris_data['Species'].value_counts()
```
