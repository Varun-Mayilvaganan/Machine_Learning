This is highly imbalanced datset and it denotes the;
0=> legit
1=> fradulent


```python
credit_card_data = credit_card_data.dropna(how='any')
```

```python
credit_card_data
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data.shape
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling due to imbalanced data

```python
# number of fraudulent transactions --> 2
legit_sample = legit.sample(n=2)
print(legit_sample.shape)
```

```python
# concatenate two data points
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```
